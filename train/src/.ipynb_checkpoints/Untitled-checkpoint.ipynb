```lua
dofile('import.lua')
model = torch.load('MPII_4stacked.t7')
-- model = torch.load('mpii_base.t7')
```

```lua
threshold_nodes, container_nodes = model:findModules('cudnn.SpatialConvolution')
for i = 1, #threshold_nodes do
  -- Search the container for the current threshold node
  for j = 1, #(container_nodes[i].modules) do
    if container_nodes[i].modules[j] == threshold_nodes[i] then
      -- Replace with a new instance
      if container_nodes[i].modules[j].nOutputPlane == 16 then
                container_nodes[i].modules[j] = cudnn.SpatialConvolution(256, 15, 1, 1):cuda()
                container_nodes[i].modules[j].train = True
      end
    end
  end
end
```

```lua
-- dofile('import.lua')
-- model = torch.load('mpii_base.t7')
model:replace(function(module)
   if torch.typename(module) == 'cudnn.SpatialConvolution' and module.nOutputPlane == 16 then
      return cudnn.SpatialConvolution(256, 14, 1, 1):cuda()
   elseif torch.typename(module) == 'cudnn.SpatialConvolution' and module.nInputPlane == 16 then
      return cudnn.SpatialConvolution(14, 256, 1, 1):cuda()
   else
      return module
   end
end)

torch.save('mpii_base_15.t7', model)
```
